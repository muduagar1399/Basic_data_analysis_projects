```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("dark")
```

## Scatterplot

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.lmplot(
    x='sepal_length',
    y='sepal_width',
    data=iris,
    fit_reg=False,  # No regression line
    hue='species'   # Color points by species
)
```

# boxplot

```python
plt.figure(figsize=(10,8)) 
sns.boxplot(data=iris)
```

## Removing Target Variable ---species

```python
dataframe_1 = iris.drop(['species'], axis=1)
plt.figure(figsize=(10, 8))
sns.boxplot(data=dataframe_1)
```

# Color Platte 

```python
species_colors = ['#78C850',  # Setosa
                    '#F08030',  # Versicolor
                    '#6890F0',  # Verginica
                  ]
```

# Vilion plot 

```python
plt.figure(figsize=(10,8)) 
sns.violinplot( data=iris, 
               palette=species_colors) # Set color palette
```

## Swarm Plot

```python
plt.figure(figsize=(10,8)) 
species_colors = {"setosa": "red", "versicolor": "blue", "virginica": "green"}
plt.figure(figsize=(10, 8))
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors)
```

## Combining plot

```python

species_colors = {"setosa": "red", "versicolor": "blue", "virginica": "green"}

# Create the figure
plt.figure(figsize=(10, 8))

# Violin plot
sns.violinplot(
    x='species', 
    y='petal_length', 
    data=iris, 
    inner=None,  # Remove the bars inside the violins
    palette=species_colors
)

# Swarm plot
sns.swarmplot(
    x='species', 
    y='petal_length', 
    data=iris, 
    color='k',  # Make points black
    alpha=0.7   # Add slight transparency
)

# Set the title
plt.title('Combined Plot')

# Show the plot
plt.show()
```

## Correlation 

```python
corr = iris.corr()
```

```python
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr)
```

# Histogram

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=iris, x='petal_length', kde=True)
```

## Bar plot

```python
species_colors = {"setosa": "red", "versicolor": "blue", "virginica": "green"}
plt.figure(figsize=(10, 8))
sns.countplot(x='species', data=iris, palette=species_colors)
plt.xticks(rotation=-45)
plt.title('Count Plot of Species')
plt.show()
```

## catplot

```python

g=sns.catplot(
    x='species', 
    y='petal_width', 
    data=iris, 
    hue='species',  # Color by species
    kind='swarm'    # Swarmplot
)
g.set_xticklabels(rotation=-45)

# Set the title
g.fig.suptitle('Swarmplot of Petal Width by Species', y=1.02)

# Show the plot
plt.show()
```

## Density Plot

```python

plt.figure(figsize=(10, 8))
sns.kdeplot(
    x='sepal_length', 
    y='sepal_width', 
    data=iris, 
    fill=True 
)

plt.title('KDE Plot of Sepal Length vs Sepal Width')
plt.show()
```

## Joint Distribution Plot

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris)
plt.show()
```
